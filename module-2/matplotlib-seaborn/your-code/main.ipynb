#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here


x = np.linspace(-2, 2, 400)


y_x_squared = x**2
y_e_to_x = np.exp(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))


ax1.plot(x, y_x_squared)
ax1.plot(x, y_e_to_x)
ax1.set_title('Normal Scale')


ax2.plot(x, y_x_squared)
ax2.plot(x, y_e_to_x)
ax2.set_yscale('log') 
ax2.set_title('Logarithmic Scale (y-axis)')



plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python


x = np.linspace(-2, 2, 400)


y_x_squared = x**2
y_e_to_x = np.exp(x)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))


ax1.plot(x, y_x_squared, label='$y = x^2$')
ax1.plot(x, y_e_to_x, label='$y = e^x$')
ax1.set_title('Normal Scale')
ax1.legend()

ax2.plot(x, y_x_squared, label='$y = x^2$')
ax2.plot(x, y_e_to_x, label='$y = e^x$')
ax2.set_yscale('log')
ax2.set_title('Logarithmic Scale (y-axis)')
ax2.legend()


plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
import calendar

median_steps_per_month = fitbit.groupby('Months_encoded')['Steps'].mean()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 4))


ax1.bar(median_steps_per_month.index, median_steps_per_month, label='Average Steps')
ax1.set_title('Average Steps by Month')
ax1.set_xlabel("Months")
ax1.set_ylabel("Average Steps")
ax1.legend()
plt.sca(ax1)
plt.xticks(rotation=20)  


work_steps_per_month = fitbit.groupby('Work_or_Weekend')['Steps'].mean()


ax2.bar(work_steps_per_month.index, work_steps_per_month, label='Sum Steps')  
ax2.set_title('Sum Steps: Workday vs Weekend')
ax2.set_xticks([0, 1])
ax2.set_xlabel("Work or Weekend")
ax2.set_ylabel("Sum Steps")
ax2.legend()


plt.show()



```

#### Use seaborn to show the distribution of column `Age`.

```python
# your code here
sns.set_theme(style="white")
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic['Rounded_Age'] = titanic['age'].dropna().round().astype(int)


age_counts = titanic['Rounded_Age'].value_counts().sort_index().reset_index()
age_counts.columns = ['Rounded_Age', 'Count']

ax = sns.barplot(x="Rounded_Age", y="Count", data=age_counts)


ax.set_title('Count of Passengers by Rounded Age')
ax.set_xlabel('Rounded Age')
ax.set_ylabel('Number of Passengers')


ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.show()
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# your code here
titanic['Gender'] = titanic['sex']


gender_counts = titanic['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']


fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(gender_counts['Gender'], gender_counts['Count'])


ax.set_title('Count of Passengers by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Number of Passengers')

plt.show()
```

```python
# Method 2 - seaborn
# your code here
titanic['Gender'] = titanic['sex']


age_counts = titanic['Gender'].value_counts().sort_index().reset_index()
age_counts.columns = ['Gender', 'Count']

ax = sns.barplot(x="Gender", y="Count", data=age_counts)


ax.set_title('Count of Passengers by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Number of Passengers')


ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.show()
```
