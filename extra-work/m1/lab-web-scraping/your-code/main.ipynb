#### Download, parse (using BeautifulSoup), and print the content from the Trending Developers page from GitHub:

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/developers'
````

````python
# your code here
response = requests.get(url)
html = response.content
html
````

````python
parsed_html = bs4.BeautifulSoup(html, "html.parser")
parsed_html
````

````python
article = parsed_html.find("h1")
````

````python
article = parsed_html.find_all("h1", {"class":"h3 lh-condensed"})
article
````

````python
articles = []
for a in article: 
    a = str(a)
    soup_a = bs4.BeautifulSoup(a, "html.parser")
    list_a = soup_a.find_all("a")
    for e in list_a:
        articles.append(e.string.strip())
print(len(articles))
````

#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url_imagelinks = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
# your code here
response_imagelinks = requests.get(url_imagelinks)
html_imagelinks = response_imagelinks.content
html_imagelinks
````

````python
parsed_imagelinks = bs4.BeautifulSoup(html_imagelinks, "html.parser")
parsed_imagelinks
````

````python
imagelinks_1 = parsed_imagelinks.find_all("img")
imagelinks_1
````

````python
imagelinks = []
for a in imagelinks_1: 
    a = str(a)
    soup_a = bs4.BeautifulSoup(a, "html.parser")
    img_tags = soup_a.find_all("img")
    for img in img_tags:
        src_attribute = img.get("src")
        if src_attribute:
            imagelinks.append(src_attribute.strip())
print(len(imagelinks))
imagelinks
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
response = requests.get(url)


if response.status_code == 200:
    
    soup = BeautifulSoup(response.text, "html.parser")

    
    language_table = soup.find("table", {"class": "wikitable"})

    
    rows = language_table.find_all("tr")[1:11]  

    
    languages = []
    native_speakers = []

    
    for row in rows:
        columns = row.find_all("td")
        if len(columns) >= 4:
            language = columns[1].text.strip()
            native_speaker = columns[2].text.strip()
            languages.append(language)
            native_speakers.append(native_speaker)

    
    data = {"Language": languages, "Native Speakers": native_speakers}
    df = pd.DataFrame(data)

    
    display(df)

else:
    print("Failed to retrieve the Wikipedia page.")
````

#### Find the book name, price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
for e in soup.find_all("article"):
    book_names = []
    prices = []
    stock_availability = []


    book_name = e.h3.a['title']
    book_names.append(book_name)

price = e.select('div p.price_color')[0].get_text()
prices.append(price)

stock = e.select('div p.availability')[0].get_text().strip()
stock_availability.append(stock)

book_data = pd.DataFrame({
            'Book Name': book_names,
            'Price': prices,
            'Stock Availability': stock_availability
        })

book_data.head()
#Unable to scrape more than one
````
